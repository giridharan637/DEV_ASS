# **1.Line Plot**

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[10,20,25,30,40]
plt.plot(x,y)
plt.title("Line Plot")
plt.show()
```

Explanation:Line plot shows data change over time.
It is useful for trend analysis.
Points are connected using straight lines.
It helps identify increase or decrease patterns.
Mostly used in time series data.

# **2.Bar Plot**

```python
students=['A','B','C','D']
marks=[80,65,90,75]
plt.bar(students,marks)
plt.title("Bar Plot")
plt.show()
```

Explanation:
Bar plot compares different categories.
Values are represented using bars.
Height of bar shows data value.
Useful for comparing student marks or sales.
Easy to understand category differences.

# **3.Histogram**

```python
data=[10,20,20,30,40,40,40,50]
plt.hist(data)
plt.title("Histogram")
plt.show()
```

Explanation:
Histogram shows frequency distribution.
Data is grouped into intervals.
It helps understand data spread.
Useful in statistics and analysis.
Shows how often values occur.

# **4.Scatter Plot**

```python
x=[1,2,3,4,5]
y=[5,7,8,9,11]
plt.scatter(x,y)
plt.title("Scatter Plot")
plt.show()
```

Explanation:
Scatter plot shows relationship between two variables.
Each point represents one observation.
Useful to find correlation between data.
Helps detect patterns and clusters.
Widely used in data science.

# **5.Pie Chart**

```python
labels=['Python','Java','C++']
sizes=[50,30,20]
plt.pie(sizes,labels=labels,autopct='%1.1f%%')
plt.title("Pie Chart")
plt.show()
```

Explanation:
Pie chart shows percentage distribution.
Data is divided into slices.
Each slice represents category share.
Useful in population or market share analysis.
Easy visualization of proportions.

# **6.Box Plot**

```python
data=[10,12,15,20,25,30,100]
plt.boxplot(data)
plt.title("Box Plot")
plt.show()
```

Explanation:
Box plot shows data spread.
It displays median and quartiles.
Outliers can be easily identified.
Useful in statistical data analysis.
Shows data variation clearly.

# **7.Area Plot**

```python
x=[1,2,3,4]
y=[3,6,2,5]
plt.fill_between(x,y)
plt.title("Area Plot")
plt.show()
```

Explanation:
Area plot shows cumulative data.
Area under line represents value.
Useful for growth visualization.
Helps compare total quantities.
Similar to line plot but filled.


# **8.Stem Plot**

```python
x=[1,2,3,4]
y=[5,3,7,9]
plt.stem(x,y)
plt.title("Stem Plot")
plt.show()
```

Explanation:
Stem plot displays discrete data points.

# **9.Step Plot**

```python
x=[1,2,3,4]
y=[10,20,15,25]
plt.step(x,y)
plt.title("Step Plot")
plt.show()
```

Explanation:
Step plot shows data in steps.

# **10.Violin Plot**

```python
data=[10,20,30,40,50]
plt.violinplot(data)
plt.title("Violin Plot")
plt.show()
```

Explanation:
Violin plot shows data distribution and density.

# **11.Bubble Chart**

```python
import matplotlib.pyplot as plt
x=[10,20,30,40,50]
y=[5,15,25,35,45]
size=[100,200,300,400,500]
plt.scatter(x,y,s=size)
plt.title("Bubble Chart")
plt.show()
```

Explanation:
Bubble chart shows relationship between three variables.
It is an advanced form of scatter plot.
X and Y axis show two variables.
Bubble size represents third variable value.
It is useful in data comparison and analysis.

# **12.Scatter Plot using Seaborn**

```python
import seaborn as sns
import matplotlib.pyplot as plt

x=[10,20,30,40,50]
y=[5,15,25,35,45]

sns.scatterplot(x=x,y=y)
plt.title("Scatter Plot using Seaborn")
plt.show()
```

Explanation:
Scatter plot shows relationship between two variables.
Seaborn provides better and attractive visualization.
It helps identify correlation and patterns.
Different colors can represent categories.
Mostly used in data analysis and machine learning.

# **13.Table Chart**

```python
import matplotlib.pyplot as plt

data=[['A',80],['B',70],['C',90],['D',85]]
column_labels=['Student','Marks']

plt.table(cellText=data,colLabels=column_labels,loc='center')
plt.axis('off')
plt.title("Table Chart")
plt.show()
```

Explanation:
Table chart displays data in tabular form.
It shows rows and columns clearly.
Useful for presenting structured data.
Helps compare values easily.
Mostly used in reports and summaries.

# **14.Polar Plot**

```python
import matplotlib.pyplot as plt
import numpy as np

theta=np.linspace(0,2*np.pi,100)
r=theta

plt.polar(theta,r)
plt.title("Polar Plot")
plt.show()
```

Explanation:
Polar plot represents data using angles and radius.
It is useful for circular or periodic data.
Data is plotted in polar coordinate system.
Commonly used in physics and engineering fields.
Helps visualize directional or rotational data.
